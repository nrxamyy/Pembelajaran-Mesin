```python
df = pd.read_csv('/content/Busiest-European-Airports-2021.csv')

# Membuat bar chart
plt.figure(figsize=(25, 12))
colors = ['#BEE9F0' for _ in range(len(df))]
colors[-69] = '#41BCD2'
plt.bar(df['Rank2021'], df['Airport'], color=colors)  # Memperbaiki nama kolom jika perlu
plt.title('Busiest Europe Airports 2021')
plt.xlabel('Airport of Europe')
plt.ylabel('Rank')
plt.xticks(rotation=90, ha='right', fontsize=12)
plt.yticks(range(0,101,20))
plt.tight_layout()
plt.show()
```
